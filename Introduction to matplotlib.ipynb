## Two options for subplot

```python
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
# Plot each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_btr_price.keys(), nut_btr_price.values());
ax4.hist(np.random.randn(1000));
```

## option 2

```python
fig, ax =plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))

#plot to each different index
ax[0, 0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_btr_price.keys(),nut_btr_price.values());
ax[1,1].hist(np.random.randn(1000))
```

## Which one should use? (pyplot vs matplotlib OO method?)
  * when ploting something quickly, okay to use pyplot method 
  * when plotting something more advanced, use the oo method

```python
heart_disease.head()
```

```python
over_50=  heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50.head())
```

```python
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
## oo method mixed with pyplot
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim(45,100)
```

```python
darde_dil=pd.read_csv("11.2 heart-disease.csv")
```

```python
darde_dil
```

```python
darde_dil.plot(x="age",y="chol",c='target',kind='scatter')

```

```python
over_50=darde_dil[darde_dil["age"]>50]
```

```python
## oo method from scratch

fig,ax= plt.subplots(figsize=(10,6))

scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#customize the plot

ax.set(title="darde_dil",
      xlabel="Age",
      ylabel="Cholestrol")

#add legend

ax.legend(*scatter.legend_elements(),title="target");

# add a horzontalline
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
darde_dil.head(6)
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))
#add data to ax0

scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
#customize ax0
ax0.set(title="darde_dil_part2",
       xlabel="age",
       ylabel="cholesterol")
#Add a legend to ax0

ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline

ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--')
# add data 2 ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])
#customize ax1
ax1.set(title="dil_dhadkan",
       xlabel="Age",
       ylabel="thalach")
#add legend to ax1
ax1.legend(*scatter.legend_elements(),title="target")
#add line to plot
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--')

# Add a title to the figure
fig.suptitle("Heart_bimari analysis", fontsize=16, fontweight="bold")
```
